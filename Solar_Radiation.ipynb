```python
x = df.drop(['Radiation'],axis=1)
y = df['Radiation'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train,y_train=scaler.fit_transform(x_train),scaler.fit_transform(y_train)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
dt=DecisionTreeRegressor()
lr=LinearRegression()
rf=RandomForestRegressor()
xgb=XGBRegressor()
knn=KNeighborsRegressor()
estimators=[('lr',lr),('rf',rf),('xgb',xgb),('dt',dt),('knn',knn)]
```

```python
vote = VotingRegressor(estimators=estimators)
```

```python
vote.fit(x_train,x_test)
```

```python
vote.score(x_train,x_test)
```

```python
vote.score(y_train,y_test)
```
